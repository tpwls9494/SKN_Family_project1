### 통계 관련 numpy

#### 중심 경향 측도

```python
import numpy as np
height = np.array([160, 170, 180, 190, 150, 150, 165, 160, 173, 169, 177, 158, 162, 155, 172, 180])
```

```python
# 평균값
print(np.mean(height))
```

```python
# 중위값
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값(데이터에 자주 등장하는 값)
values, count = np.unique(height, return_counts=True)
print(values, count)

max_count = np.max(height)
print(count == max_count)
print(values[count == max_count])

for i in range(len(values)):
    if count[i] == np.max(count):
        print(values[i], end=" ")
```

### 산포 측도

```python
# 분산 variance
# (값 - 평균)^2 의 평균
var1 = np.mean(np.square(height - np.mean(height)))
var2 = np.var(height)
print(var1)
print(var2)
```

```python
# 표준편차 standard deviation
# 분산의 제곱근
str1 = np.sqrt(var1)
str2 = np.std(var2)
print(str1)
print(str2)
```

```python
# 범위

print(np.max(height) - np.min(height))
```

### 확률분포

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 정규분포
arr = np.random.randn(1000)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 정규분포
# np.random.normal(loc, scale, size)
# - loc : 평균
# - scale : 표준편차
# - size : 갯수 or shape
arr = np.random.normal(10, 1 ,100)

plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.normal(loc, scale, size)
# - loc : 평균
# - scale : 표준편차
# - size : 갯수 or shape
arr = np.random.normal(10, 1 ,(10, 100))

plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.rand(num)
# - 0.0 이상 ~ 1.0 미만 수에서 균등하게 난수 추출
# - num : 갯수 or shape
np.random.rand(1000)
np.random.rand(2, 3)

plt.hist(arr, bins=100)
plt.show()
```

```python
arr = np.random.randint(1, 100, 1000)
print(arr)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등분포
# np.random.randint(start, end, size)
# - start 이상 end 미만의 난수(정수)를 size만큼 추출
np.random.seed(0)   # 난수 계산 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)
print(arr)

plt.hist(arr, bins=100)
plt.show()
```
