### 최근접이웃 분류 모델로 훈련/평가 해보기

```python
# 데이터 전처리
# iris_data = load_iris()
# iris_data.keys()

# 1. 입력-라벨 데이터
train_input, train_label, test_input, testlabel = list(map(iris_data.split()))

mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)

# 3. 정규화
train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std
```

```python
# 데이터 전처리

# 1. 입력-라벨 데이터

# 2. 훈련-테스트 데이터
from sklearn.model_selection import train_test_split
train_input, test_input, train_label, test_label = train_test_split(iris_data.data, iris_data.target, test_size=25, random_state=42)
train_input.shape, test_input.shape, train_label.shape, test_label.shape

# 3. 정규화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_input)
train_scaled = scaler.transform(train_input)
test_scaled = scaler.transform(test_input)
# test_scaled
```

```python
# 훈련
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)
```

```python
# 평가
kn.score(test_scaled, test_label)
```

```python
# 평가 : accuracy_score
from sklearn.metrics import accuracy_score
pred = kn.predict(test_scaled)
accuracy_score(test_label, pred)
```
